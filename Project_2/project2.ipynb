Project Introduction: Smart City Citizen Activity Analysis
1. Dataset Overview

This dataset, smart_city_citizen_activity.csv, is a rich synthetic dataset designed to simulate the activities, habits, and environmental impact of citizens within a modern smart city. It contains 2,000 anonymized records, each representing a unique citizen, with 15 distinct attributes that quantify various aspects of their daily lives.

The primary purpose of this dataset is to enable data-driven analysis for urban planning, public health initiatives, sustainability programs, and the optimization of city services. By exploring the relationships between these variables, one can derive actionable insights to improve citizen well-being and reduce the city's environmental footprint.
2. Features Description

The features can be broadly categorized into four groups:

A. Demographic Information:

    Citizen_ID: A unique identifier for each citizen.

    Age: The age of the citizen.

    Gender: The gender of the citizen (Male, Female, Other).

B. Daily Activity & Lifestyle:

    Work_Hours: Number of hours spent working.

    Shopping_Hours: Number of hours spent shopping.

    Entertainment_Hours: Number of hours spent on entertainment.

    Social_Media_Hours: Number of hours spent on social media.

    Public_Events_Hours: Number of hours spent attending public events.

    Sleep_Hours: Number of hours slept.

C. Health & Mobility:

    Mode_of_Transport: Primary mode of transportation (e.g., Car, EV, Walking, Bicycle, Bike, Public Transport).

    Steps_Walked: The number of steps walked during the day.

    Calories_Burned: Estimated calories burned.

D. Energy & Environmental Impact:

    Home_Energy_Consumption_kWh: Energy consumption at home in kilowatt-hours.

    Charging_Station_Usage: Binary indicator (0 or 1) if the citizen used an EV charging station.

    Carbon_Footprint_kgCO2: The total carbon footprint in kilograms of CO2 equivalent. This is a key performance indicator (KPI) for sustainability.

3. Potential Problems for EDA (Project Ideas)

This dataset is ideal for exploring a wide range of questions. Here are some compelling directions for your analysis:

    Sustainable Transportation Analysis:

        How does the choice of Mode_of_Transport directly influence an individual's Carbon_Footprint_kgCO2?

        What is the demographic profile (Age, Gender) of citizens who use electric vehicles (EVs) or bicycles?

        Do citizens who walk or cycle more (Steps_Walked) have better health metrics (Calories_Burned, Sleep_Hours)?

    Lifestyle and Health Correlation:

        Is there a observable trade-off between Work_Hours/Social_Media_Hours and Sleep_Hours or Calories_Burned?

        Can we identify clusters of citizens based on their activity patterns (e.g., "active commuters," "sedentary workers," "social butterflies")?

    Energy Consumption Patterns:

        What factors (Age, Work_Hours, Home_Energy_Consumption_kWh) are most predictive of a high Carbon_Footprint_kgCO2?

        Is there a correlation between time spent at home (inversely related to Work_Hours + Public_Events_Hours) and Home_Energy_Consumption_kWh?

    Demographic Studies:

        How do lifestyles and carbon footprints differ across age groups and genders?

        For example, do younger citizens use more social media and have different transportation habits than older citizens?

4. Technical Approach for EDA

Your analysis should follow a structured EDA workflow:

    Data Cleaning: Check for missing or inconsistent values.

    Univariate Analysis: Understand the distribution of each variable (e.g., histograms, boxplots for Age, Carbon_Footprint).

    Bivariate Analysis: Explore relationships between pairs of variables (e.g., scatter plots of Steps_Walked vs. Calories_Burned, boxplots of Carbon_Footprint by Mode_of_Transport).

    Multivariate Analysis: Create insights using grouping and segmentation (e.g., average carbon footprint by Mode_of_Transport and Age group).

    Correlation Analysis: Identify strong positive or negative correlations between numerical variables.

IMPORTANT QUESTIONS

 What is the average carbon footprint by mode of transport?

```python
avg_carbon = df.groupby('Mode_of_Transport')['Carbon_Footprint_kgCO2'].mean().sort_values()
avg_carbon.plot(kind='bar', title='Average Carbon Footprint by Transport Mode')
plt.ylabel('kgCO2')
plt.show()
```

How does age distribution look?

```python
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.title('Age Distribution of Citizens')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Is there a correlation between steps walked and calories burned?

```python
import seaborn as sns

sns.scatterplot(data=df, x='Steps_Walked', y='Calories_Burned')
plt.title('Steps Walked vs Calories Burned')
plt.show()
```

Which gender uses which transport mode most?

```python
cross_tab = pd.crosstab(df['Gender'], df['Mode_of_Transport'])
cross_tab.plot(kind='bar', stacked=True)
plt.title('Transport Mode Usage by Gender')
plt.show()
```

What is the average sleep duration by gender?

```python
df.groupby('Gender')['Sleep_Hours'].mean().plot(kind='bar')
plt.title('Average Sleep Hours by Gender')
plt.ylabel('Hours')
plt.show()
```

Do EV users have lower carbon footprints?

```python
ev_users = df[df['Mode_of_Transport'] == 'EV']
non_ev = df[df['Mode_of_Transport'] != 'EV']

print("EV Users Avg Carbon:", ev_users['Carbon_Footprint_kgCO2'].mean())
print("Non-EV Users Avg Carbon:", non_ev['Carbon_Footprint_kgCO2'].mean())
```

 How does home energy consumption relate to carbon footprint?

```python
sns.scatterplot(data=df, x='Home_Energy_Consumption_kWh', y='Carbon_Footprint_kgCO2')
plt.title('Home Energy vs Carbon Footprint')
plt.show()
```

What is the distribution of work hours?

```python
plt.hist(df['Work_Hours'], bins=10, edgecolor='black')
plt.title('Distribution of Work Hours')
plt.xlabel('Work Hours')
plt.show()
```

 Which age group has the highest carbon footprint?

```python
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 50, 70, 100], labels=['Young', 'Middle', 'Senior', 'Elderly'])
df.groupby('Age_Group')['Carbon_Footprint_kgCO2'].mean().plot(kind='bar')
plt.title('Avg Carbon Footprint by Age Group')
plt.show()
```

 How many people use charging stations?

```python
charging_usage = df['Charging_Station_Usage'].value_counts()
charging_usage.plot(kind='pie', autopct='%1.1f%%')
plt.title('Charging Station Usage')
plt.show()
```

What is the average carbon footprint for each gender?

```python
df.groupby('Gender')['Carbon_Footprint_kgCO2'].mean().plot(kind='bar')
plt.title('Average Carbon Footprint by Gender')
plt.ylabel('kgCO2')
plt.show()
```

. Is there a relationship between social media hours and sleep?

```python
sns.scatterplot(data=df, x='Social_Media_Hours', y='Sleep_Hours')
plt.title('Social Media Hours vs Sleep Hours')
plt.show()
```

 Which transport mode is most popular?

```python
df['Mode_of_Transport'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Popularity of Transport Modes')
plt.show()
```

What is the average carbon footprint for EV vs Car users?

```python
car_ev = df[df['Mode_of_Transport'].isin(['Car', 'EV'])]
car_ev.groupby('Mode_of_Transport')['Carbon_Footprint_kgCO2'].mean().plot(kind='bar')
plt.title('Carbon Footprint: Car vs EV')
plt.show()
```

How does carbon footprint vary with age

```python
plt.scatter(df['Age'], df['Carbon_Footprint_kgCO2'], alpha=0.5)
plt.title('Carbon Footprint vs Age')
plt.xlabel('Age')
plt.ylabel('Carbon Footprint (kgCO2)')
plt.show()
```
